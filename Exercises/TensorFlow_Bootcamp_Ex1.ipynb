```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the boston. dataset from tensorflow
(X_train, y_train), (x_test, y_test) = tf.keras.datasets.boston_housing.load_data(path="boston_housing.npz", 
                                                                                  test_split=0.2, 
                                                                                  seed=42)
```

```python
print(X_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Create a neural network regression model

```python
# set the random seed
tf.random.set_seed(42)

# Create the model
boston_model = tf.keras.Sequential([
          tf.keras.layers.Dense(25),
          tf.keras.layers.Dense(1)
])

#2. compile the model
boston_model.compile(loss = tf.keras.losses.mae,
                     optimizer = tf.keras.optimizers.Adam(),
                     metrics = ["mae"])

# 3. fit the model
boston_model.fit(X_train, y_train, epochs = 100)
```

```python
boston_model.evaluate(x_test, y_test)
```
