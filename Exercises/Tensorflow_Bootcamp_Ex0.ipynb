###1. Create a vector, scalar, matrix and tensor with values of your choosing using `tf.constant()`

```python
import tensorflow as tf
```

```python
# Scalar 
sc = tf.constant(4)
sc
```

```python
sc.ndim
```

```python
# Vector
vec = tf.constant([3,5,4,2,7])
vec
```

```python
vec.ndim
```

```python
# Matrix
mat = tf.constant([[10, 7],
                   [3,5]])
mat
```

```python
mat.ndim, tf.size(mat), mat.shape
```

```python
# Tensor -> Matrix of higher dimension
tens = tf.constant([[[10, 4, 3],
                     [4, 5, 6],
                     [2,1,7]],
                    
                    [[3, 5, 6],
                     [6, 7, 3],
                     [5, 3, 2]]])

tens
```

```python
tens.ndim, tf.size(tens), tens.shape
```

### 2. Create two tensors containing random values between 0 and 1 with shape `[5, 300]`.

```python
tf.random.set_seed(42)
rand_tens1 = tf.random.uniform(shape = [5, 300], seed = 42)
rand_tens1
```

```python
tf.random.set_seed(10)
rand_tens2 = tf.random.uniform(shape = [5, 300], seed = 10)
rand_tens2
```

### 3. Multiply the two tensors you created in 2 using matrix multiplication.

```python
tf.matmul(rand_tens1, tf.transpose(rand_tens2))
```

### 4. Multiply the two tensors you created in 2 using dot product.

```python
tf.tensordot(rand_tens1, tf.transpose(rand_tens2), axes = 1)
```

###5. Create a tensor with random values between 0 and 1 with shape `[224, 224, 3]`.

```python
tens_shape_3 = tf.random.uniform(shape = (224, 224, 3))
tens_shape_3
```

### 6. Find the min and max values of the tensor you created in 5 along the first axis.


```python
tf.reduce_max(tens_shape_3, axis=0)
```

```python
tf.reduce_max(tens_shape_3, axis=1)
```

```python
tf.reduce_max(tens_shape_3, axis=2)
```

### 7. Created a tensor with random values of shape `[1, 224, 224, 3]` then squeeze it to change the shape to `[224, 224, 3]`.

```python
# New tensor
A = tf.random.uniform(shape = (1,224,224,3))
A
```

```python
squeeze_A = tf.squeeze(A)
squeeze_A
```

### 8. Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value.

```python
# Single dimension tensor
tf.random.set_seed(23)
single_tensor = tf.constant([4,5,3,2,1,6,7,8,9,10])
single_tensor
```

```python
tf.argmax(single_tensor)
```

### 9. One-hot encode the tensor you created in 8
